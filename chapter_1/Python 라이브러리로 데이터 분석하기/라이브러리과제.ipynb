```python
import pandas as pd
```

```python
#Q1
df_excel = pd.read_excel('관서별 5대범죄 발생 및 검거.xlsx')    # 엑셀에서 데이터프레임 읽어오기
print(df_excel)    # 데이터프레임 출력
```

```python

#Q2
pol_list = {'서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',     # 딕셔너리 값을 저장
'서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
'혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
'마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
'광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
'강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
'구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
'방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'
}

df_excel["구별"] = df_excel["관서명"].map(pol_list).fillna("구 없음")   # 딕셔너리 값을 이용해 "구별"이라는 새로운 열을 관서명에 대응해 추가

print(df_excel)   # 데이터 프레임 출력
```

```python
#Q3
df_excel = pd.pivot_table(df_excel, index='관서명', aggfunc='sum')    # 관서명을 인덱스로 지정하고 중복되는 값을 더해주기
df_excel = df_excel.set_index("구별")   # "구별"열을 인덱스로 지정해서 "관서명"을 탈락시킴
```

```python
#Q4
df_excel = df_excel.drop(labels='구 없음', axis=0)    # 구 없음 행 삭제
```

```python
#Q5
a = ["강간", "강도", "살인", "절도", "폭력", "소계"]    # 검거율을 만들기 위해 접두사 리스트 만들기
for word in a:    # 접두사 리스트를 for문에 넣어주기
    if word != "소계":
        df_excel[f"{word}검거율"] = df_excel[f"{word}(검거)"] / df_excel[f"{word}(발생)"] * 100    # "소계" 이외의 값일 때 실행해 계산한 값을 "{word}검거율"로 데이터프레임에 추가
    elif word == "소계":
        df_excel["검거율"] = df_excel[f"{word}(검거)"] / df_excel[f"{word}(발생)"] * 100    # "소계" 값일 때 실행해 계산한 값을 "검거율"로 데이터프레임에 추가
```

```python
#Q6
del_1 = ["강간(검거)","강도(검거)","살인(검거)","절도(검거)","폭력(검거)", "소계(발생)","소계(검거)"]   # 삭제할 열을 리스트 값으로 만들기
for word_1 in del_1:    # for 문에 del_1 넣어주기
    del df_excel[word_1]    # 열 삭제
```

```python
#Q7
df_excel.rename(columns = lambda x : x.replace("(발생)", ""), inplace = True)     # rename()으로 replace() 메서드를 사용해 "(발생)" 부분을 ""로 치환해서 "(발생)"부분을 제거
```

```python
#추가도전과제 1
df_csv = pd.read_csv('pop_kor.csv')     # csv 데이터프레임 읽어오기
df_csv.set_index("구별", inplace = True)    # 인덱스 값을 "구별"로 지정하고 원본값 수정
print(df_csv)    # 데이터프레임 출력
```

```python
#추가도전과제 2
df_excel = df_excel.join(df_csv) # join()을 사용해 데이터프레임에 열 추가하기
print(df_excel) # 데이터프레임 출력
```

```python
#추가도전과제 3
df_excel = df_excel.sort_values(by='검거율') # "검거율" 열을 기준으로 데이터프레임 오름차순 정렬
print(df_excel) # 데이터프레임 출력
```
